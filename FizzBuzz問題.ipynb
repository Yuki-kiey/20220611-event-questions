```python
for num in range(1, 101):
  string = ''
  #15の倍数→15で割ってあまりが0の場合
  if num%15 == 0:
    string = 'FizzBuzz'
  #3の倍数→15の倍数ではないが、3で割ってあまりが0の場合
  elif num%3 == 0:
    string = 'Fizz'
  #5の倍数→15の倍数ではないが、5で割ってあまりが0の場合
  elif num%5 == 0:
    string = 'Buzz'
  else:
    string = num
  print(string)
```
